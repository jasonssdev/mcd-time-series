# Autocorrelación y Autocorrelación Parcial
**¿Qué es la autocorrelación?**
La autocorrelación es una medida que nos indica cuánto se correlaciona una serie de tiempo consigo misma en diferentes desplazamientos temporales. Es decir, ¿cuánto se parece una serie de tiempo a sí misma si la movemos
k periodos en el tiempo?

**¿Qué es la autocorrelación parcial?**
Mientras que la función de autocorrelación nos da la correlación total entre dos puntos de una serie de tiempo separados por
k periodos, la función de autocorrelación parcial nos da la correlación entre esos dos puntos después de eliminar la correlación de todos los puntos intermedios.

#Importando Librerías

```python
from darts import TimeSeries
from darts.datasets import AirPassengersDataset
from darts.utils.statistics import check_seasonality, plot_acf, plot_pacf
import matplotlib.pyplot as plt
```

```python
air_passengers_series = AirPassengersDataset().load()
air_passengers_series.plot()
```

#Autocorrelación

```python
plot_acf(air_passengers_series)
plt.title("Autocorrelation Function (ACF) of Air Passengers")
plt.xlabel("Lags")
plt.ylabel("Autocorrelation")
plt.show()
```

Decaimiento lento y amortiguado → fuerte autocorrelación: La autocorrelación no cae rápido a cero, como en un MA.
Clara periodicidad → estacionalidad: el pico fuerte se repite cada 12 rezagos, picos en lag 12, 24, 36.

# Autocorrelación Parcial

```python
plot_pacf(air_passengers_series)
plt.title("Partial Autocorrelation Function (PACF) of Air Passengers")
plt.xlabel("Lags")
plt.ylabel("Partial Autocorrelation")
plt.show()
```
