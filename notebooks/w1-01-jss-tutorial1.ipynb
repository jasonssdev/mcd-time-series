```python
import pandas as pd
from darts import TimeSeries
import matplotlib.pyplot as plt
from scipy.stats import boxcox
import numpy as np
from darts.datasets import AirPassengersDataset
```

```python
dataset = AirPassengersDataset()
series = dataset.load()
```

```python
series.plot()
plt.show()
```

```python
# Trandormacion Logaritmica

log_data = np.log(series.to_series())
log_series = TimeSeries.from_series(log_data)

log_series.plot()
plt.show()
```

```python
# Transformacion - Diferenciacion
diff_series = series.diff()
diff_series.plot()
plt.show()
```

```python
# Box Cox

values = series.to_series().values
boxcox_values, lambda_value = boxcox(values)
print(f'Lambda: {lambda_value}')
```

```python
boxcox_series = TimeSeries.from_values(boxcox_values)
boxcox_series.plot()
plt.show()
```
